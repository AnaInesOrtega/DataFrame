### Retraso de vuelo EDA (Análisis exploratorio de datos)

Acerca del conjunto de datos Contexto La Oficina de Estadísticas de Transporte del Departamento de Transporte de EE. UU. (DOT) realiza un seguimiento de la puntualidad de los vuelos nacionales operados por grandes compañías aéreas. La información resumida sobre la cantidad de vuelos a tiempo, retrasados, cancelados y desviados se publica en el Informe mensual de consumidores de viajes aéreos del DOT y en este conjunto de datos de retrasos y cancelaciones de vuelos de 2008.

Se realizará un análisis del conjunto de datos proporcionado, este contiene hasta 1.936.758 vuelos internos diferentes en los EE. UU. para 2008 y sus causas de retraso, desvío y cancelación. Los datos provienen de la Oficina de Estadísticas de Transporte (BTS) del Departamento de Transporte de EE. UU. (DOT). Los resultados son tentativos.

Este conjunto de datos está compuesto por las siguientes variables: 
1.	Year: 2008.
2.	Month: 1-12.
3.	DayofMonth: 1-31.  
4.	DayOfWeek: 1 (lunes) - 7 (domingo). 
5.	DepTime: hora de salida real (local, hhmm).
6.	CRSDepTime: hora de salida programada (local, hhmm). 
7.	ArrTime: hora de llegada real (local, hhmm). 
8.	CRSArrTime: hora de llegada programada (local, hhmm). 
9.	UniqueCarrier: código único de la compañía aérea.
10.	FlightNum: número de vuelo. 
11.	TailNum: número de cola del avión: matrícula de la aeronave, identificador único de la aeronave.
12.	ActualElapsedTime: en minutos.
13.	CRSElapsedTime: en minutos.
14.	AirTime: en minutos.  
15.	ArrDelay: demora en la llegada, en minutos: Un vuelo se cuenta como "a tiempo" si se realizó con menos de 15 minutos de retraso con respecto a la hora programada que se muestra en los Sistemas de Reservas Computarizados (CRS) de las aerolíneas. 
16.	DepDelay: retraso en la salida, en minutos. 
17.	Origin: código de origen del aeropuerto IATA. 
18.	Dest: código de destino de aeropuerto IATA.
19.	Distance: en millas.
20.	TaxiIn: taxi a tiempo en minutos.
21.	TaxiOut: taxi fuera de tiempo en minutos. 
22.	Cancelled: *era el vuelo cancelado.
23.	CancellationCode: motivo de la cancelación (A = transportista, B = clima , C = NAS, D = seguridad). 
24.	Diverted:1 = sí, 0 = no. 
25.	CarrierDelay: en minutos: El retraso del transportista está bajo el control del transportista aéreo. Ejemplos de sucesos que pueden determinar el retraso del transportista son: limpieza de la aeronave, daños en la aeronave, espera de la llegada de pasajeros o tripulantes de conexión, equipaje, colisión con aves, carga de carga, catering, computadora, interrupción del equipo del transportista, legalidad de la tripulación (descanso del piloto o asistente) , daños por mercancías peligrosas, inspección de ingeniería, abastecimiento de combustible, manejo de pasajeros discapacitados, tripulación retrasada, servicio de lavabos, mantenimiento, sobreventas, servicio de agua potable, retiro de pasajeros ingobernables, embarque o asiento lentos, almacenamiento de equipaje de mano, retrasos por peso y equilibrio. 
26.	WeatherDelay: en minutos: el retraso del tiempo es causado por condiciones climáticas extremas o peligrosas que se pronostican o se manifiestan en el punto de partida, en ruta o en el punto de llegada.
27.	NASDelay: en minutos: el retraso que está bajo el control del Sistema Nacional del Espacio Aéreo (NAS) puede incluir: condiciones climáticas no extremas, operaciones aeroportuarias, volumen de tráfico pesado, control de tráfico aéreo, etc. 
28.	SecurityDelay: en minutos: el retraso de seguridad es causado por la evacuación de una terminal o vestíbulo, volver a abordar una aeronave debido a una violación de la seguridad, equipos de control inoperativos y/o largas filas de más de 29 minutos en las áreas de proyección.
29.	LateAircraftDelay: en minutos: Retraso en la llegada a un aeropuerto debido a la llegada tardía de la misma aeronave a un aeropuerto anterior. El efecto dominó de un retraso anterior en los aeropuertos posteriores se denomina propagación del retraso.

### Eliminamos filas con datos nulos

```python
retraso_vuelos.dropna(inplace=True)
retraso_vuelos.shape
```

### Crea columnes noves (velocitat mitjana del vol)

```python
retraso_vuelos['Distance_Km'] = retraso_vuelos['Distance'] * 1.60934 # conversión millas a km
retraso_vuelos['AirTime'].replace(0, np.nan, inplace = True)
retraso_vuelos['AirTime_H'] = retraso_vuelos['AirTime'] / 60 #conversión minutos a h.
retraso_vuelos['Average_Speed'] = retraso_vuelos['Distance_Km'] / retraso_vuelos['AirTime_H'] #columna velocidad media
retraso_vuelos[['Distance', 'AirTime', 'Average_Speed']].head()
```

```python
retraso_vuelos['Average_Speed'].mean()
```

### Crea columnas nuevas (si ha llegado tarde)

```python
retraso_vuelos['Late_Landing'] = retraso_vuelos['ArrDelay'] > 0
retraso_vuelos[['UniqueCarrier','Late_Landing']].head(10)
```

### Crea columnas nuevas (si ha salido tarde)

```python
retraso_vuelos['Late_Take_Off'] = retraso_vuelos['DepDelay'] > 0
retraso_vuelos[['Distance', 'AirTime', 'Average_Speed', 'Late_Landing', 'Late_Take_Off']].head(10)
```

### Tabla de las aerolínies con más retrasos acumulados.

```python
retrasos_acumulados = retraso_vuelos['Late_Landing'] == True
retraso_vuelos[retrasos_acumulados]['UniqueCarrier'].value_counts()
```

### Vuelos más largos

```python
vuelos_largos = retraso_vuelos.sort_values(by= ['Distance'], ascending = False)
vuelos_largos[:10] # muestra los 10 primeros vuelos más largos
```

### Vuelos más retrasados 

```python
vuelos_mas_retrasados = retraso_vuelos.sort_values(by = ['ArrDelay'], ascending = False) #ordeno los vuelos en forma descendente y muestro los 20 vuelos más retrasados
print(vuelos_mas_retrasados[:20]) 
```
